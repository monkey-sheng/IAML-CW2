```python
# Using the training data of class 0 (’A’) from the training set Xtrn_m
# calculate the sample mean vector, and unbiased sample covariance matrix using Numpy’s functions
class0_m = Xtrn_m[Ytrn == 0]
print(class0_m.shape)
# (a) min,max,mean of elems of covariance matrix
cov_matrix = np.cov(class0_m.T, ddof=1)
print(cov_matrix.shape)
print(cov_matrix.min(), cov_matrix.max(), cov_matrix.mean())
# print([cov_matrix[i,i] for i in range(0,300)])
diagonal_elems = cov_matrix.diagonal()
print(diagonal_elems.min(), diagonal_elems.max(), diagonal_elems.mean())
```

```python
# (c)
fig,ax=plt.subplots(1,1,figsize=(10,10))
ax.hist(diagonal_elems,bins=15)
ax.grid()
ax.set_title('Diagonal elements of covariance matrix histogram')
ax.set_xlabel('cov values')
```

```python
from scipy.stats import multivariate_normal as mulnorm
mean_vector = Xtrn_m.mean(axis=0)
gaussian_dist = mulnorm(mean_vector, cov_matrix)
gaussian_dist.pdf(Xtst_m[0])
```

```python
# 2.8 (a)
from sklearn.mixture import GaussianMixture
gm = GaussianMixture(n_components=1, covariance_type='full')
# Use { Xtrn_m, Ytrn } as the training set and { Xtst_m, Ytst} as the test set.
gm.fit(class0_m)
gm.score_samples(Xtst_m[Ytst==0])[0]
```

```python
# 2.8 (b)
from sklearn.naive_bayes import GaussianNB
gms = []
for letter in range(26):
    class_letter = Xtrn_m[Ytrn==letter]
    gm = GaussianMixture()
    gm.fit(class_letter, Ytrn[Ytrn==letter])
    gms.append(gm)

print(Xtst_m.shape)

preds = np.array([gm.score_samples(Xtst_m) for gm in gms])
print(preds.shape)
misclassification_count = 0
for i in range(Xtst_m.shape[0]):
    pred_letter = np.argmax(preds[:,i])
    if pred_letter != Ytst[i]:
        misclassification_count += 1
print('testing set accuracy')
print(Xtst_m.shape[0], misclassification_count)

preds = np.array([gm.score_samples(Xtrn_m) for gm in gms])
misclassification_count = 0
for i in range(Xtrn_m.shape[0]):
    pred_letter = np.argmax(preds[:,i])
    if pred_letter != Ytrn[i]:
        misclassification_count += 1
print('training set accuracy')
print(Xtrn_m.shape[0], misclassification_count)
```
